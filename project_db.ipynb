```python


%%sql

CREATE OR REPLACE FUNCTION fn_login_with_username_and_password(
    p_username TEXT,
    p_password TEXT
)
RETURNS INT AS
$$
DECLARE
    v_user_id INT;
BEGIN
    -- Check if there is a user with this username + password
    SELECT user_id
    INTO v_user_id
    FROM user_account
    WHERE username = p_username
      AND password = p_password;

    -- If no user found → login failed
    IF v_user_id IS NULL THEN
        RETURN 1;  -- login failed
    END IF;

    -- If a user is found → login success
    RETURN 0;
END;
$$ LANGUAGE plpgsql;


```

```python
%%sql
SELECT fn_login_with_username_and_password('owner1', '1234');
```
